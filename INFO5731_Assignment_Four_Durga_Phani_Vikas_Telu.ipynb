```python
import nltk
review_data['clean_text'] = review_data['clean_text'].str.replace('[^\w\s]','')
review_data['clean_text'] = review_data['clean_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))#lower case
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
review_data['clean_text'] = review_data['clean_text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#keywords
tfidf= TfidfVectorizer(ngram_range=(1,2), max_features=1000)
tfidf.fit(review_data['clean_text'])
x_values =  tfidf.transform(review_data['clean_text'])
y_values = review_data['sentiment']

from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_values, y_values, test_size = 0.2)
```

```python
from sklearn import metrics

def evaluation(y_pred, y_test):
    Acc = metrics.accuracy_score(y_pred, y_test)
    recall = metrics.recall_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    Prec = metrics.precision_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    F1_score= 2 * (Prec * recall) / (Prec + recall) # Formula for F1 Score
    print("Accuracy: ", Acc.round(4))
    print("Recall:", recall.round(4))
    print("Precision:", Prec.round(4))
    print("F-1 score:", F1_score.round(4))
```

```python
from sklearn import naive_bayes
from sklearn.model_selection import cross_val_score, KFold

nb_implement = naive_bayes.MultinomialNB()

nb_implement.fit(x_train,y_train)

y_pred_valid = nb_implement.predict(x_valid)
evaluation(y_pred_valid, y_valid)
print("Cross Validation Score:", cross_val_score(nb_implement, x_valid, y_valid, cv= KFold(10, shuffle=True, random_state = 22)))
```

```python
from sklearn import svm

svm_imp = svm.SVC()

svm_imp.fit(x_train,y_train)

y_pred_valid = svm_imp.predict(x_valid)
evaluation(y_pred_valid, y_valid)
print("Cross Validation Score:", cross_val_score(svm_imp, x_valid, y_valid, cv= KFold(10, shuffle=True, random_state = 22)).mean().round(4))
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, RandomSampler, SequentialSampler, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np


custom_model_name = "QCRI/bert-base-multilingual-cased-pos-english"
custom_tokenizer = BertTokenizer.from_pretrained(custom_model_name)
custom_model = BertForSequenceClassification.from_pretrained(custom_model_name)

custom_df = pd.read_csv('/content/kgf2_imdb_reviews_sent.csv')


# Preprocess function
def custom_preprocess(clean_text, sentiment, tokenizer, max_len=512):
    input_ids, attention_masks, labels_out = [], [], []

    for clean_text, sentiment in zip(clean_text, sentiment):
        # Convert label to int (0 or 1)
        sentiment = 0 if sentiment == 'negative' else 1

        encoded_dict = tokenizer.encode_plus(
            clean_text,
            add_special_tokens=True,
            max_length=max_len,
            padding='max_length',  # Updated padding argument
            truncation=True,       # Enable truncation
            return_attention_mask=True,
            return_tensors='pt'
        )

        input_ids.append(encoded_dict['input_ids'])
        attention_masks.append(encoded_dict['attention_mask'])
        labels_out.append(sentiment)

    input_ids = torch.cat(input_ids, dim=0)
    attention_masks = torch.cat(attention_masks, dim=0)
    labels_out = torch.tensor(labels_out)

    return input_ids, attention_masks, labels_out


```

```python

custom_train_texts, custom_test_texts, custom_train_labels, custom_test_labels = train_test_split(custom_df['clean_text'], custom_df['sentiment'], test_size=0.3)

custom_train_inputs, custom_train_masks, custom_train_labels = custom_preprocess(custom_train_texts, custom_train_labels, custom_tokenizer)
custom_test_inputs, custom_test_masks, custom_test_labels = custom_preprocess(custom_test_texts, custom_test_labels, custom_tokenizer)

custom_batch_size = 16
custom_train_data = TensorDataset(custom_train_inputs, custom_train_masks, custom_train_labels)
custom_train_sampler = RandomSampler(custom_train_data)
custom_train_dataloader = DataLoader(custom_train_data, sampler=custom_train_sampler, batch_size=custom_batch_size)

custom_test_data = TensorDataset(custom_test_inputs, custom_test_masks, custom_test_labels)
custom_test_sampler = SequentialSampler(custom_test_data)
custom_test_dataloader = DataLoader(custom_test_data, sampler=custom_test_sampler, batch_size=custom_batch_size)



```

```python
def custom_evaluate(dataloader, model):
    model.eval()
    predictions, true_labels = [], []

    for batch in dataloader:
        b_input_ids, b_input_mask, b_labels = batch
        with torch.no_grad():
            outputs = model(b_input_ids, token_type_ids=None, attention_mask=b_input_mask)

        logits = outputs.logits
        logits = logits.detach().cpu().numpy()
        label_ids = b_labels.to('cpu').numpy()

        predictions.extend(logits)
        true_labels.extend(label_ids)

    predictions = np.array(predictions)
    true_labels = np.array(true_labels)

    pred_flat = np.argmax(predictions, axis=1).flatten()
    labels_flat = true_labels.flatten()
    return accuracy_score(labels_flat, pred_flat)

custom_accuracy = custom_evaluate(custom_test_dataloader, custom_model)
print(custom_accuracy)
```
